### plotly express를 이용하여 라인 차트 시각화 

```python
query = """
select date_trunc('day', order_date)::date as day
	, sum(amount) as sum_amount, count(distinct a.order_id) as daily_ord_cnt
from nw.orders a
	join nw.order_items b on a.order_id = b.order_id
group by date_trunc('day', order_date)::date 
order by 1
"""
df = pd.read_sql_query(sql=query, con=postgres_engine)
df.head(10)
```

```python
import plotly.express as px
fig = px.line(data_frame=df, x='day', y='sum_amount')
fig.show()
```

### plotly graph_objects를 이용하여 라인 차트 시각화

```python
import plotly.graph_objects as go

fig = go.Figure()
# 라인 차트는 go.Scatter()를 호출
fig.add_trace(go.Scatter(
    x=df['day'],
    y=df['sum_amount'],
    name='daily amount'
))
fig.show()
```

### plotly express를 이용하여 바 차트 시각화

```python
query = """
select date_trunc('month', order_date)::date as month
	, sum(amount) as sum_amount, count(distinct a.order_id) as daily_ord_cnt
from nw.orders a
	join nw.order_items b on a.order_id = b.order_id
group by date_trunc('month', order_date)::date 
order by 1
"""
df = pd.read_sql_query(sql=query, con=postgres_engine)
df.head(10)
```

```python
import plotly.express as px
fig = px.bar(data_frame=df, x='month', y='sum_amount')
fig.show()
```

### plotly graph_objects를 이용하여 바 차트 시각화

```python
import plotly.graph_objects as go

fig = go.Figure()
# 라인 차트는 go.Scatter()를 호출
fig.add_trace(go.Bar(
    x=df['month'],
    y=df['sum_amount'],
    name='monthly amount'
))
fig.show()
```
